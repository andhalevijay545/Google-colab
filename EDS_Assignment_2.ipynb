```python
import numpy as np
import pandas as pd
FILE_PATH = "IPL2025Batters.csv"
df = pd.read_csv(FILE_PATH)
print("=" * 65)
print("        IPL 2025 BATTERS — NUMPY & PANDAS ASSIGNMENT")
print("=" * 65)
print(f"\nDataset loaded successfully.  Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}\n")

#PROBLEM STATEMENT 1
# Find the total number of runs scored by all batters in IPL 2025
# using NumPy sum.
print("─" * 65)
print("PROBLEM 1: Total runs scored by all batters")
print("─" * 65)
runs_array = df["Runs"].to_numpy()
total_runs = np.sum(runs_array)
print(f"Total Runs Scored (NumPy) : {total_runs}\n")

#PROBLEM STATEMENT 2
# Display the top 5 run-scorers of IPL 2025 using Pandas.
print("─" * 65)
print("PROBLEM 2: Top 5 run-scorers of IPL 2025")
print("─" * 65)
top5 = df.nlargest(5, "Runs")[["Player Name", "Team", "Runs"]]
print(top5.to_string(index=False), "\n")

#PROBLEM STATEMENT 3
# Calculate the mean, median, and standard deviation of batting
# averages using NumPy.
print("─" * 65)
print("PROBLEM 3: Mean, Median & Std Dev of Batting Averages")
print("─" * 65)
# Replace '-' with NaN before converting to float
df['AVG'] = df['AVG'].replace('-', np.nan)
df['AVG'] = pd.to_numeric(df['AVG']) # Explicitly convert to numeric after replacement
avg_array = df["AVG"].to_numpy(dtype=float) # Convert to float
print(f"Mean   AVG : {np.nanmean(avg_array):.2f}") # Use nanmean
print(f"Median AVG : {np.nanmedian(avg_array):.2f}") # Use nanmedian
print(f"Std Dev AVG: {np.nanstd(avg_array):.2f}\n") # Use nanstd

# PROBLEM STATEMENT 4
# Find all batters who scored more than 400 runs using Pandas
# boolean filtering.
print("─" * 65)
print("PROBLEM 4: Batters who scored more than 400 runs")
print("─" * 65)
over_400 = df[df["Runs"] > 400][["Player Name", "Team", "Runs"]]
print(over_400.to_string(index=False), "\n")

#PROBLEM STATEMENT 5
# Find the player with the highest strike rate (SR) using Pandas
# idxmax().
print("─" * 65)
print("PROBLEM 5: Player with the highest Strike Rate")
print("─" * 65)
idx = df["SR"].idxmax()
best_sr = df.loc[idx, ["Player Name", "Team", "SR"]]
print(best_sr.to_string(), "\n")

#PROBLEM STATEMENT 6
# Count the number of players representing each team using Pandas
# value_counts().
print("─" * 65)
print("PROBLEM 6: Number of players per team")
print("─" * 65)
team_counts = df["Team"].value_counts()
print(team_counts.to_string(), "\n")

#PROBLEM STATEMENT 7
# Calculate total 4s and 6s hit by each team using Pandas groupby
# and sum.
print("─" * 65)
print("PROBLEM 7: Total 4s and 6s hit by each team")
print("─" * 65)
team_boundaries = df.groupby("Team")[["4s", "6s"]].sum().reset_index()
team_boundaries["Total Boundaries"] = team_boundaries["4s"] + team_boundaries["6s"]
team_boundaries = team_boundaries.sort_values("Total Boundaries", ascending=False)
print(team_boundaries.to_string(index=False), "\n")

#PROBLEM STATEMENT 8
# Use NumPy to find the batter with the maximum and minimum number
# of balls faced (BF).
print("─" * 65)
print("PROBLEM 8: Batter with max and min balls faced")
print("─" * 65)
bf_array = df["BF"].to_numpy()
max_idx = np.argmax(bf_array)
min_idx = np.argmin(bf_array)
print(f"Most  Balls Faced : {df.loc[max_idx, 'Player Name']} — {bf_array[max_idx]} balls")
print(f"Least Balls Faced : {df.loc[min_idx, 'Player Name']} — {bf_array[min_idx]} balls\n")

#PROBLEM STATEMENT 9
# Compute the correlation between Runs and Strike Rate (SR) using
# NumPy corrcoef.
print("─" * 65)
print("PROBLEM 9: Correlation between Runs and Strike Rate")
print("─" * 65)
corr = np.corrcoef(df["Runs"].to_numpy(), df["SR"].to_numpy())
print(f"Correlation (Runs vs SR) : {corr[0, 1]:.4f}")
print("(Values close to 1 or -1 indicate strong correlation)\n")

#PROBLEM STATEMENT 10
# Add a new column 'Boundary_Runs' that stores runs scored from
# boundaries (4s × 4 + 6s × 6) using Pandas
print("─" * 65)
print("PROBLEM 10: Add Boundary_Runs column (4s×4 + 6s×6)")
print("─" * 65)
df["Boundary_Runs"] = df["4s"] * 4 + df["6s"] * 6
print(df[["Player Name", "4s", "6s", "Boundary_Runs"]].head(10).to_string(index=False), "\n")

#PROBLEM STATEMENT 11
# Find the team with the highest average batting average (AVG) using
# Pandas groupby and mean.
print("─" * 65)
print("PROBLEM 11: Team with the highest average batting average")
print("─" * 65)
team_avg = df.groupby("Team")["AVG"].mean().sort_values(ascending=False)
print(team_avg.to_string())
print(f"\nBest Team (Avg AVG): {team_avg.idxmax()} — {team_avg.max():.2f}\n")

#PROBLEM STATEMENT 12
# Use NumPy to create a boolean array identifying batters who hit
# at least one century (100s >= 1) and display their names.
print("─" * 65)
print("PROBLEM 12: Batters with at least one century (NumPy mask)")
print("─" * 65)
hundreds_array = df["100s"].to_numpy()
mask = hundreds_array >= 1
centurions = df["Player Name"].to_numpy()[mask]
print(f"Total Centurions : {mask.sum()}")
print("Players          :", list(centurions), "\n")

#PROBLEM STATEMENT 13
# Sort all players by Strike Rate (SR) in descending order and
# display the top 10 using Pandas sort_values.
print("─" * 65)
print("PROBLEM 13: Top 10 batters by Strike Rate (descending)")
print("─" * 65)
top_sr = df.sort_values("SR", ascending=False).head(10)[["Player Name", "Team", "SR"]]
print(top_sr.to_string(index=False), "\n")

#PROBLEM STATEMENT 14
# Calculate the percentage of runs scored from boundaries vs total
# runs for each player and add it as a new column using Pandas.
print("─" * 65)
print("PROBLEM 14: Boundary Run % for each player")
print("─" * 65)
df["Boundary_%"] = ((df["Boundary_Runs"] / df["Runs"]) * 100).round(2)
print(df[["Player Name", "Runs", "Boundary_Runs", "Boundary_%"]].head(10).to_string(index=False), "\n")

#PROBLEM STATEMENT 15
# Use NumPy to compute the 25th, 50th and 75th percentile of runs
# scored (IQR analysis).
print("─" * 65)
print("PROBLEM 15: Percentile analysis of Runs (25th, 50th, 75th)")
print("─" * 65)
runs_np = df["Runs"].to_numpy()
p25 = np.percentile(runs_np, 25)
p50 = np.percentile(runs_np, 50)
p75 = np.percentile(runs_np, 75)
iqr  = p75 - p25
print(f"25th Percentile : {p25:.2f}")
print(f"50th Percentile : {p50:.2f}")
print(f"75th Percentile : {p75:.2f}")
print(f"IQR (P75 - P25) : {iqr:.2f}\n")

#PROBLEM STATEMENT 16
# Find players who played all their innings without getting out
# even once (Not Outs == Inn) using Pandas filtering.
print("─" * 65)
print("PROBLEM 16: Players never dismissed (No == Inn)")
print("─" * 65)
never_out = df[df["No"] == df["Inn"]][["Player Name", "Team", "Inn", "No", "Runs"]]
if never_out.empty:
    print("No player remained not-out in every innings.")
else:
    print(never_out.to_string(index=False))
print()

#PROBLEM STATEMENT 17
# Rank players within each team by their runs using Pandas
# groupby + rank.
print("─" * 65)
print("PROBLEM 17: Rank players within each team by Runs")
print("─" * 65)
df["Team_Rank"] = df.groupby("Team")["Runs"].rank(method="dense", ascending=False).astype(int)
print(df[["Player Name", "Team", "Runs", "Team_Rank"]]
      .sort_values(["Team", "Team_Rank"])
      .head(20)
      .to_string(index=False), "\n")

#PROBLEM STATEMENT 18
# Use NumPy to normalise (min-max scaling) the Runs column to a
# 0–1 range and display the first 10 values
print("─" * 65)
print("PROBLEM 18: Min-Max Normalisation of Runs column")
print("─" * 65)
runs_np = df["Runs"].to_numpy(dtype=float)
norm_runs = (runs_np - np.min(runs_np)) / (np.max(runs_np) - np.min(runs_np))
df["Normalised_Runs"] = np.round(norm_runs, 4)
print(df[["Player Name", "Runs", "Normalised_Runs"]].head(10).to_string(index=False))

#PROBLEM STATEMENT 19
# Display a summary of the dataset using Pandas describe() to
# understand the distribution of numerical columns.
print("─" * 65)
print("PROBLEM 19: Statistical Summary using describe()")
print("─" * 65)
summary = df[["Runs", "Matches", "Inn", "AVG", "SR", "4s", "6s"]].describe().round(2)
print(summary.to_string(), "\n")

#PROBLEM STATEMENT 20
# Identify all-round performers — batters with above-average Runs
# AND above-average Strike Rate — using NumPy logical operations.
print("─" * 65)
print("PROBLEM 20: All-round performers (above avg Runs AND above avg SR)")
print("─" * 65)
runs_np = df["Runs"].to_numpy()
sr_np   = df["SR"].to_numpy()

mean_runs = np.mean(runs_np)
mean_sr   = np.mean(sr_np)

# Create boolean masks
above_avg_runs = runs_np > mean_runs
above_avg_sr   = sr_np   > mean_sr

# Combine masks using logical AND
all_rounders_mask = np.logical_and(above_avg_runs, above_avg_sr)

# Filter player names using the combined mask
all_rounders = df["Player Name"].to_numpy()[all_rounders_mask]

print(f"Total All-Rounders : {all_rounders_mask.sum()}")
print("Players            :", list(all_rounders))
```
